```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage.transform import resize

# --- Cell 1: Load and Preprocess Image ---
image = io.imread('https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png')
gray_image = color.rgb2gray(image)
gray_image = resize(gray_image, (256, 256), anti_aliasing=True)
gray_image = (gray_image * 255).astype(np.uint8)
original_image_float = gray_image.astype(np.float32)

# --- Cell 2: SVD Functions ---
def compress_block(block, k):
    U, S, VT = np.linalg.svd(block, full_matrices=False)
    S_k = np.diag(S[:k])
    U_k = U[:, :k]
    VT_k = VT[:k, :]
    return np.dot(U_k, np.dot(S_k, VT_k))

def compress_image_svd(image, k):
    h, w = image.shape
    compressed = np.zeros_like(image, dtype=np.float32)
    for i in range(0, h, 8):
        for j in range(0, w, 8):
            block = image[i:i+8, j:j+8]
            compressed_block = compress_block(block, k)
            compressed[i:i+8, j:j+8] = compressed_block
    return np.clip(compressed, 0, 255).astype(np.uint8)

# --- Cell 3: Analysis Functions ---
def compression_ratio(k):
    original = 64
    compressed = k * (8 + 8 + 1)
    return original / compressed

def reconstruction_error(original, compressed):
    return np.linalg.norm(original - compressed, 'fro')

def psnr(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if mse == 0:
        return float('inf')
    max_pixel = 255.0
    return 20 * np.log10(max_pixel / np.sqrt(mse))

# --- Cell 4: Run Experiments ---
errors = []
ratios = []
psnrs = []
reconstructed_images = {}

for k in range(1, 9):
    rec_img = compress_image_svd(original_image_float, k)
    err = reconstruction_error(original_image_float, rec_img)
    cr = compression_ratio(k)
    p = psnr(original_image_float, rec_img)

    errors.append(err)
    ratios.append(cr)
    psnrs.append(p)
    reconstructed_images[k] = rec_img

    plt.imsave(f'svd_k_{k}.png', rec_img, cmap='gray')

# --- Cell 5: Plot Compression Ratio & Error ---
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(range(1, 9), ratios, marker='o')
plt.title('Compression Ratio vs k')
plt.xlabel('k')
plt.ylabel('Compression Ratio')

plt.subplot(1, 2, 2)
plt.plot(range(1, 9), errors, marker='x', color='red')
plt.title('Reconstruction Error vs k')
plt.xlabel('k')
plt.ylabel('Frobenius Norm')

plt.tight_layout()
plt.show()

# --- Cell 6: (Optional) Plot PSNR ---
plt.figure()
plt.plot(range(1, 9), psnrs, marker='s', color='green')
plt.title('PSNR vs k')
plt.xlabel('k')
plt.ylabel('PSNR (dB)')
plt.grid(True)
plt.show()

# --- ✅ Cell 7: Visualization - Image Comparison ---
print("\n--- Displaying Image Comparison ---")

if 'original_image_float' not in locals() or not reconstructed_images:
    print("Original or reconstructed images not found. Please run Cells 2 and 4.")
else:
    plt.figure(figsize=(16, 8))

    # Display Original Image
    plt.subplot(2, 4, 1)
    plt.imshow(original_image_float, cmap='gray')
    plt.title("Original")
    plt.axis('off')

    # Display Reconstructed Images
    k_to_display = [1, 2, 3, 4, 5, 6, 8]
    plot_position = 2
    for k_disp in k_to_display:
        if k_disp in reconstructed_images:
            plt.subplot(2, 4, plot_position)
            plt.imshow(reconstructed_images[k_disp], cmap='gray')
            plt.title(f"Reconstructed (k={k_disp})")
            plt.axis('off')
            plot_position += 1
        else:
            print(f"Warning: Reconstructed image for k={k_disp} not found.")

    plt.tight_layout()
    plt.show()

print("\n=== Jupyter Notebook Cells Finished ===")
```
