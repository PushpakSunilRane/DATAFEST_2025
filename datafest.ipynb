```python
# --- Revised Cell 16 ---
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("--- Preparing Data for Market Analysis ---")

# --- Define relevant columns based on leases_df.info() ---
market_col = 'market'
sqft_col = 'leasedSF'
cbd_suburban_col = 'CBD_suburban'
# -----------------------------------------------------------

# Check if columns exist
if not all(col in leases_df.columns for col in [market_col, sqft_col, cbd_suburban_col]):
    print(f"ERROR: One or more required columns ({market_col}, {sqft_col}, {cbd_suburban_col}) not found.")
    print(f"Available columns: {leases_df.columns.tolist()}")
else:
    print(f"Using columns: '{market_col}', '{sqft_col}', '{cbd_suburban_col}'")

    # Select relevant columns and create a copy
    market_analysis_df = leases_df[[market_col, sqft_col, cbd_suburban_col]].copy()

    # --- Clean Market Column ---
    # Fill missing market values (if any) and assign back
    market_analysis_df[market_col] = market_analysis_df[market_col].fillna('Unknown')
    # Standardize: lowercase and strip whitespace
    market_analysis_df[market_col] = market_analysis_df[market_col].astype(str).str.lower().str.strip()

    # --- Clean leasedSF Column ---
    # Ensure numeric, coerce errors to NaN
    market_analysis_df[sqft_col] = pd.to_numeric(market_analysis_df[sqft_col], errors='coerce')
    # Drop rows where leasedSF is missing or became NaN
    rows_before_drop = len(market_analysis_df)
    market_analysis_df.dropna(subset=[sqft_col], inplace=True) # dropna inplace is usually fine
    rows_after_drop = len(market_analysis_df)
    if rows_before_drop > rows_after_drop:
        print(f"Dropped {rows_before_drop - rows_after_drop} rows with missing/non-numeric '{sqft_col}'.")

    # --- Clean CBD_suburban Column ---
    # Fill missing values (if any) and assign back
    market_analysis_df[cbd_suburban_col] = market_analysis_df[cbd_suburban_col].fillna('Unknown')
    # Standardize: lowercase and strip whitespace
    market_analysis_df[cbd_suburban_col] = market_analysis_df[cbd_suburban_col].astype(str).str.lower().str.strip()

    print("\nCleaned unique values (sample):")
    print(f"Markets (sample): {market_analysis_df[market_col].unique()[:10]}")
    print(f"CBD/Suburban: {market_analysis_df[cbd_suburban_col].unique()}")

    print("\nPrepared DataFrame for market analysis:")
    display(market_analysis_df.head())
    market_analysis_df.info()
```

```python
# Re-run Cell 17 if needed to ensure the summary DataFrames are available
# Check if market_analysis_df exists
if 'market_analysis_df' in locals():
    print("\nAggregating data by market...")

    # Group by the cleaned market column and calculate metrics
    market_summary = market_analysis_df.groupby(market_col).agg(
        total_leases = pd.NamedAgg(column=market_col, aggfunc='size'),
        total_sqft = pd.NamedAgg(column=sqft_col, aggfunc='sum'),
        average_sqft = pd.NamedAgg(column=sqft_col, aggfunc='mean'),
        median_sqft = pd.NamedAgg(column=sqft_col, aggfunc='median') # Median can be robust to outliers
    ).reset_index()

    # Sort by total leases
    market_summary_sorted_leases = market_summary.sort_values(by='total_leases', ascending=False)

    # Sort by total square footage <--- We will use this one to define TOP markets
    market_summary_sorted_sqft = market_summary.sort_values(by='total_sqft', ascending=False)

    print("\nSummary by Market (Sorted by Lease Count):")
    display(market_summary_sorted_leases.head(15)) # Show top 15

    print("\nSummary by Market (Sorted by Total SqFt):")
    display(market_summary_sorted_sqft.head(15)) # Show top 15

else:
    print("ERROR: Please run the preparation cell (Cell 16) successfully first.")
```

```python
# --- Revised Cell 19 ---
# Check if market_summary_sorted_sqft exists
if 'market_summary_sorted_sqft' in locals():
    top_n = 15 # Number of top markets to show
    print(f"\nVisualizing Average Lease Size (SqFt) in Top {top_n} Markets (Ranked by Total SqFt)...") # Updated print statement

    # --- Select top N markets based on TOTAL SQFT ---
    top_markets_by_sqft_list = market_summary_sorted_sqft[market_col].head(top_n).tolist()
    # Filter the summary data for these top markets
    plot_data_avg_sqft = market_summary[market_summary[market_col].isin(top_markets_by_sqft_list)]
    # Sort this subset by average_sqft for better visualization
    plot_data_avg_sqft = plot_data_avg_sqft.sort_values('average_sqft', ascending=False)


    plt.figure(figsize=(10, 7))
    sns.barplot(data=plot_data_avg_sqft, y=market_col, x='average_sqft', palette='coolwarm', orient='h')
    plt.title(f'Average Lease Size (SqFt) in Top {top_n} Markets (Ranked by Total SqFt)') # Updated title
    plt.xlabel('Average Square Footage per Lease')
    plt.ylabel('Market')
    plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x/1e3:.0f}K'))
    plt.tight_layout()
    plt.show()

    # --- Box Plot for Distribution (More detailed view) ---
    print(f"\nVisualizing Lease Size Distribution (Box Plot) for Top {top_n} Markets (Ranked by Total SqFt)...") # Updated print statement
    # Filter the original analysis data for these top markets (using the list derived from total sqft ranking)
    boxplot_data = market_analysis_df[market_analysis_df[market_col].isin(top_markets_by_sqft_list)]

    plt.figure(figsize=(12, 8))
    # Order the boxplot by median size (descending) for clarity
    order = boxplot_data.groupby(market_col)[sqft_col].median().sort_values(ascending=False).index
    sns.boxplot(data=boxplot_data, y=market_col, x=sqft_col, palette='Spectral', orient='h', order=order, showfliers=False) # showfliers=False hides outliers
    plt.title(f'Distribution of Lease Sizes (SqFt) in Top {top_n} Markets (Ranked by Total SqFt, Outliers Hidden)') # Updated title
    plt.xlabel('Leased Square Footage (Log Scale)')
    plt.ylabel('Market')
    plt.xscale('log')
    plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x/1e3:.0f}K' if x >= 1000 else f'{x:.0f}'))
    plt.tight_layout()
    plt.show()

else:
    print("ERROR: Please run the aggregation cell (Cell 17) successfully first.")
```

```python
# --- Revised Cell 20 ---
# Check if market_analysis_df and market_summary_sorted_sqft exist
if 'market_analysis_df' in locals() and 'market_summary_sorted_sqft' in locals():
    top_n = 15 # Use the same top N markets
    # --- Select top N markets based on TOTAL SQFT ---
    top_markets_by_sqft_list = market_summary_sorted_sqft[market_col].head(top_n).tolist()
    print(f"\nAnalyzing CBD vs Suburban Activity in Top {top_n} Markets (Ranked by Total SqFt)...") # Updated print statement

    # Filter for top markets (using the list derived from total sqft ranking)
    cbd_suburban_data = market_analysis_df[market_analysis_df[market_col].isin(top_markets_by_sqft_list)]

    # Group by market and CBD/Suburban status, count leases
    cbd_summary = cbd_suburban_data.groupby([market_col, cbd_suburban_col]).size().unstack(fill_value=0)

    # Calculate proportions
    cbd_summary_props = cbd_summary.apply(lambda x: x / x.sum(), axis=1)

    # Ensure the order matches the top N list for plotting consistency
    # Reindex based on the order from the total sqft sorted list
    cbd_summary_props = cbd_summary_props.reindex(top_markets_by_sqft_list)

    print("\nProportion of Leases: CBD vs Suburban in Top Markets (Ranked by Total SqFt)") # Updated print statement
    display(cbd_summary_props)

    # --- Visualize using a stacked bar chart ---
    print(f"\nVisualizing CBD vs Suburban Proportion in Top {top_n} Markets (Ranked by Total SqFt)...") # Updated print statement
    fig, ax = plt.subplots(figsize=(12, 7))
    cbd_summary_props.plot(kind='bar', stacked=True, ax=ax, colormap='coolwarm')
    ax.set_title(f'Proportion of Leases: CBD vs. Suburban in Top {top_n} Markets (Ranked by Total SqFt)') # Updated title
    ax.set_xlabel('Market')
    ax.set_ylabel('Proportion of Total Leases')
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    plt.xticks(rotation=45, ha='right')
    plt.legend(title=cbd_suburban_col.replace('_', ' ').title())
    plt.tight_layout()
    plt.show()

else:
    print("ERROR: Please run the preparation (Cell 16) and aggregation (Cell 17) cells successfully first.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Preparing Data for Class Analysis by Market ---")

# --- Define relevant columns ---
market_col = 'market'
class_col = 'internal_class'
# -----------------------------

# Check if columns exist
if not all(col in leases_df.columns for col in [market_col, class_col]):
    print(f"ERROR: One or more required columns ({market_col}, {class_col}) not found.")
    print(f"Available columns: {leases_df.columns.tolist()}")
else:
    print(f"Using columns: '{market_col}', '{class_col}'")

    # Select relevant columns and create a copy
    market_class_analysis_df = leases_df[[market_col, class_col]].copy()

    # --- Clean Market Column ---
    market_class_analysis_df[market_col] = market_class_analysis_df[market_col].fillna('Unknown')
    market_class_analysis_df[market_col] = market_class_analysis_df[market_col].astype(str).str.lower().str.strip()

    # --- Clean Class Column ---
    market_class_analysis_df[class_col] = market_class_analysis_df[class_col].fillna('Unknown')
    market_class_analysis_df[class_col] = market_class_analysis_df[class_col].astype(str).str.lower().str.strip()

    print("\nUnique values:")
    print(f"Markets (sample): {market_class_analysis_df[market_col].unique()[:10]}")
    print(f"Classes: {market_class_analysis_df[class_col].unique()}")

    print("\nPrepared DataFrame for market-class analysis:")
    display(market_class_analysis_df.head())
    market_class_analysis_df.info()
```

```python
# Check if market_class_analysis_df exists
if 'market_class_analysis_df' in locals():
    print("\nAggregating lease count by market and building class...")

    # Group by market AND class, then count leases in each group
    market_class_summary = market_class_analysis_df.groupby([market_col, class_col]).size()

    # Convert the resulting Series to a DataFrame
    market_class_summary = market_class_summary.reset_index(name='total_leases')

    print("\nSummary of Leases by Market and Building Class (Sample):")
    display(market_class_summary.head(10))

else:
    print("ERROR: Please run the preparation cell (Cell 24) successfully first.")
```

```python
# Check if market_class_summary and top market list exist
# Make sure 'market_summary_sorted_sqft' from Cell 17 is available
if 'market_class_summary' in locals() and 'market_summary_sorted_sqft' in locals():
    top_n = 15 # Number of top markets to show (ranked by total sqft)
    # Get the list of top N markets based on total sqft
    top_markets_by_sqft_list = market_summary_sorted_sqft[market_col].head(top_n).tolist()

    print(f"\nVisualizing Lease Count by Class for Top {top_n} Markets (Ranked by Total SqFt)...")

    # Filter the summary data for only the top N markets
    plot_data = market_class_summary[market_class_summary[market_col].isin(top_markets_by_sqft_list)]

    # --- Create Grouped Bar Chart ---
    plt.figure(figsize=(14, 8)) # Wider figure for grouped bars

    # Use seaborn's barplot with hue for grouping by class
    # Order the markets based on the top_markets_by_sqft_list for consistency
    ax = sns.barplot(data=plot_data, x=market_col, y='total_leases', hue=class_col,
                     order=top_markets_by_sqft_list, palette='Set2') # Use order parameter

    plt.title(f'Lease Count by Building Class in Top {top_n} Markets (Ranked by Total SqFt)')
    plt.xlabel('Market')
    plt.ylabel('Total Leases')
    plt.xticks(rotation=45, ha='right') # Rotate labels if they overlap
    plt.legend(title=class_col.replace('_', ' ').title()) # Nicer legend title

    # Optional: Format y-axis
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'{y/1e3:.1f}K' if y >= 1000 else f'{y:.0f}'))

    plt.tight_layout()
    plt.show()

    # --- Alternative: Stacked Bar Chart (Proportions) ---
    print(f"\nVisualizing Proportion of Leases by Class for Top {top_n} Markets...")

    # Pivot the data to get markets as index, classes as columns, counts as values
    market_class_pivot = plot_data.pivot(index=market_col, columns=class_col, values='total_leases').fillna(0)

    # Calculate proportions
    market_class_props = market_class_pivot.apply(lambda x: x / x.sum(), axis=1)

    # Reindex to maintain the desired market order
    market_class_props = market_class_props.reindex(top_markets_by_sqft_list)

    # Plot stacked bar chart
    fig, ax = plt.subplots(figsize=(14, 8))
    market_class_props.plot(kind='bar', stacked=True, ax=ax, colormap='Set2') # Use same palette
    ax.set_title(f'Proportion of Leases by Building Class in Top {top_n} Markets (Ranked by Total SqFt)')
    ax.set_xlabel('Market')
    ax.set_ylabel('Proportion of Total Leases')
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y))) # Format y-axis as percentage
    plt.xticks(rotation=45, ha='right')
    plt.legend(title=class_col.replace('_', ' ').title(), bbox_to_anchor=(1.02, 1), loc='upper left') # Move legend outside
    plt.tight_layout()
    plt.subplots_adjust(right=0.85) # Adjust layout to make space for legend
    plt.show()


else:
    print("ERROR: Please run the aggregation cell (Cell 25) and ensure 'market_summary_sorted_sqft' from Cell 17 is available.")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Ensure pandas is imported

print("--- Preparing Occupancy Data for Yearly Analysis ---")

# --- Define relevant columns ---
year_col = 'year'
market_col = 'market' # Keep market for potential breakdown later
avg_occ_col = 'avg_occupancy_proportion'
# -----------------------------

# Check if occupancy_df exists first
if 'occupancy_df' not in locals():
    print("ERROR: 'occupancy_df' not found. Please load the occupancy data first.")
# Check if columns exist in occupancy_df
elif not all(col in occupancy_df.columns for col in [year_col, market_col, avg_occ_col]):
    print(f"ERROR: One or more required columns ({year_col}, {market_col}, {avg_occ_col}) not found in occupancy_df.")
    print(f"Available columns: {occupancy_df.columns.tolist()}")
else:
    print(f"Using columns: '{year_col}', '{market_col}', '{avg_occ_col}'")

    # Select relevant columns and create a copy
    occupancy_yearly_df = occupancy_df[[year_col, market_col, avg_occ_col]].copy()

    # --- Clean Market Column (for consistency) ---
    occupancy_yearly_df[market_col] = occupancy_yearly_df[market_col].fillna('Unknown')
    occupancy_yearly_df[market_col] = occupancy_yearly_df[market_col].astype(str).str.lower().str.strip()

    # --- Verify Data Types (should be correct based on info) ---
    occupancy_yearly_df[year_col] = pd.to_numeric(occupancy_yearly_df[year_col], errors='coerce')
    occupancy_yearly_df[avg_occ_col] = pd.to_numeric(occupancy_yearly_df[avg_occ_col], errors='coerce')

    # Drop rows if year or occupancy became NaN (unlikely here)
    occupancy_yearly_df.dropna(subset=[year_col, avg_occ_col], inplace=True)

    print("\nPrepared DataFrame for yearly occupancy analysis:")
    display(occupancy_yearly_df.head())
    occupancy_yearly_df.info()

```

```python
# Check if occupancy_yearly_df exists
if 'occupancy_yearly_df' in locals():
    print("\nAggregating average occupancy proportion by year (across included markets)...")

    # Group by year and calculate the mean of the average occupancy proportions
    # This averages the quarterly averages for each market within that year,
    # then averages across all markets for that year.
    yearly_avg_occupancy = occupancy_yearly_df.groupby(year_col)[avg_occ_col].mean()

    # Convert to DataFrame
    yearly_avg_occupancy = yearly_avg_occupancy.reset_index()

    print("\nAverage Relative Occupancy Proportion per Year:")
    display(yearly_avg_occupancy)
    print("\nNOTE: Occupancy is relative to March 1, 2020 baseline.")
    print("A value of 1.0 means occupancy = baseline; < 1.0 means lower than baseline.")

else:
    print("ERROR: Please run the preparation cell (Cell 36) successfully first.")

```
